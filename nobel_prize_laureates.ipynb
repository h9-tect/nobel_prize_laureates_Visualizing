# This code reads a CSV file, combines the 'Firstname' and 'Surname' columns into a new 'Fullname' column, drops unnecessary columns, reorders the remaining columns, and then prints the first few rows of the modified dataset.

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv("nobel_prize_laureates.csv", delimiter=";")

# Combine 'Firstname' and 'Surname' into 'Fullname'
data['Fullname'] = data['Firstname'] + ' ' + data['Surname']

# Drop unnecessary columns
columns_to_drop = [ 'Firstname', 'Surname', 'Born country code', 'Born city', 'Died', 'Died country', 'Died country code', 'Overall motivation', 'Organization city', 'Geo Shape', 'Geo Point 2D']
data = data.drop(columns=columns_to_drop)

# Reorder columns
data = data[['Id', 'Fullname'] + [col for col in data.columns if col not in ['Id', 'Fullname']]]

# Print the head of the dataset
print(data.head())
```

```python
# Drop rows with NaN values
data = data.dropna()
```

# isualizing the normalized distribution of Nobel Prize categories by birth country through a stacked bar chart

```python
import plotly.express as px

# Crosstab of Nobel Prize categories by birth country
cross_table = pd.crosstab(data['Born country'], data['Category'])

# Normalize the cross table by dividing each value by the sum of its row
cross_table_normalized = cross_table.div(cross_table.sum(axis=1), axis=0)

# Reset index to have 'Born country' as a column
cross_table_normalized = cross_table_normalized.reset_index()

# Melt the dataframe to have a long format for plotting
melted_df = pd.melt(cross_table_normalized, id_vars='Born country', var_name='Category', value_name='Normalized Count')

# Sort the dataframe by 'Normalized Count' in descending order
melted_df = melted_df.sort_values('Normalized Count', ascending=False)

# Create the stacked bar chart using Plotly
fig = px.bar(melted_df, y='Born country', x='Normalized Count', color='Category',
             orientation='h', barmode='stack', width=800, height=600)

# Customize the layout of the figure
fig.update_layout(
    title='Normalized Distribution of Nobel Prize Categories by Birth Country',
    xaxis_title='Normalized Count',
    yaxis_title='Birth Country',
    legend_title='Category',
    font=dict(size=12),
    margin=dict(l=150, r=50, t=80, b=50)  # Adjust margins for better visibility
)

fig.show()
```

# Visualizing the age at the time of receiving the Nobel Prize by year using a scatter plot, grouped by category

```python
import pandas as pd
import plotly.express as px

# Read the CSV file and convert the 'Year' and 'Born' columns to datetime
data = pd.read_csv("nobel_prize_laureates.csv", delimiter=";", parse_dates=['Year', 'Born'])

# Create the 'Fullname' column by combining 'Firstname' and 'Surname'
data['Fullname'] = data['Firstname'] + ' ' + data['Surname']

# Calculate the age at the time of receiving the Nobel Prize
data['Age'] = (data['Year'] - data['Born']).dt.total_seconds() / (365.25 * 24 * 60 * 60)

# Create the scatter plot using Plotly
fig = px.scatter(data, x='Year', y='Age', color='Category', hover_data=['Fullname', 'Born country'],
                 trendline='ols', title='Age at Time of Receiving Nobel Prize', width=800, height=600)

# Customize the layout of the figure
fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Age',
    legend_title='Category',
    font=dict(size=12),
    margin=dict(l=80, r=50, t=80, b=50)  # Adjust margins for better visibility
)

fig.show()
```

# Performing LDA topic modeling on the Nobel Prize laureates' motivations and overall motivations, extracting 5 topics and evaluating the model's coherence score

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim import models, corpora
from gensim.models import CoherenceModel
import pyLDAvis.gensim_models
import matplotlib.pyplot as plt

nltk.download('stopwords')
nltk.download('punkt')

# Read the CSV file and drop NaN values in 'Overall motivation' and 'Motivation' columns
data = pd.read_csv("nobel_prize_laureates.csv", delimiter=";")
data = data.dropna(subset=['Overall motivation', 'Motivation'])

# Combine 'Overall motivation' and 'Motivation' columns into a single text column
data['Text'] = data['Overall motivation'] + ' ' + data['Motivation']

# Tokenize the text and remove stopwords
stop_words = set(stopwords.words('english'))
data['Tokens'] = data['Text'].apply(lambda x: [token.lower() for token in word_tokenize(x) if token.lower() not in stop_words])

# Create a dictionary from the tokenized text
dictionary = corpora.Dictionary(data['Tokens'])

# Convert tokenized text into a bag-of-words representation
corpus = [dictionary.doc2bow(tokens) for tokens in data['Tokens']]

# Perform LDA topic modeling
lda_model = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=10)

# Print the topics and their corresponding keywords
for topic_id, topic_words in lda_model.print_topics():
    print(f'Topic ID: {topic_id}')
    print(f'Topic Keywords: {topic_words}\n')

# Compute coherence score to evaluate the model
coherence_model = CoherenceModel(model=lda_model, texts=data['Tokens'], dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()
print(f'Coherence Score: {coherence_score}')


```
